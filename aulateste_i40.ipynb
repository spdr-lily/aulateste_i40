# Aula Teste
* Cenário: Caldeira Industrial </br>
* Objetivo: Construir um pipeline funcional que simula a detecção de anomalias em motores industriais, onde a IA a vai aprender a diferença entre uma variação normal e uma situação de perigo com base em exemplos anteriores.

```python
#importação de bibliotecas
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
import time
```

```python
# 1. SIMULAÇÃO DE INGESTÃO DE DADOS (Sensor IoT)
def get_sensor_data():
    """Simula a leitura de sensores de temperatura e vibração."""
    print("LOG: Recebendo dados do broker MQTT...")
    data = {
        'timestamp': [time.time()],
        'temperatura': [np.random.uniform(60, 110)], # Celsius
        'vibracao': [np.random.uniform(2, 10)]       # mm/s
    }
    return pd.DataFrame(data)
```

```python
# 2. PIPELINE DE PROCESSAMENTO (ETL)
def data_pipeline(df):
    """Limpeza e Normalização básica."""
    # Simula tratamento de valores nulos ou outliers
    df['temp_normalizada'] = df['temperatura'] / 100
    return df
```

```python
# 3. INTEGRAÇÃO COM MODELO DE IA
# (Aqui treinamos um modelo simples para fins didáticos)
def get_trained_model():
    # Dados de treino: [Temp, Vibração] -> Falha (1) ou Não (0)
    X_train = np.array([[70, 3], [80, 4], [105, 9], [110, 8], [65, 2]])
    y_train = np.array([0, 0, 1, 1, 0])

    model = LogisticRegression()
    model.fit(X_train, y_train)
    return model
```

```python
# EXECUÇÃO DO FLUXO
print("--- Iniciando Sistema Indústria 4.0 ---")
model = get_trained_model()

while True:
    # Passo 1: Ingestão
    raw_data = get_sensor_data()

    # Passo 2: Pipeline de Dados
    processed_data = data_pipeline(raw_data)

    # Passo 3: Inferência da IA
    features = processed_data[['temperatura', 'vibracao']].values
    prediction = model.predict(features)

    status = "ALERTA: Falha Iminente!" if prediction[0] == 1 else "Normal"
    print(f"Dados: Temp={raw_data['temperatura'][0]:.2f}°C | Status: {status}")
    print("-" * 40)

    time.sleep(3) # Aguarda 3 segundos para a próxima leitura
```
